```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Input is an array
input_array = np.array([1, 2, 3])
exp_array = np.exp(input_array)

print("Input to exp:", input_array)
print("Output of exp:", exp_array)

# Input is a single value
input_val = 1
exp_val = np.exp(input_val)

print("\nInput to exp:", input_val)
print("Output of exp:", exp_val)
```

```python
def sigmoid(z):
    """
    Compute the sigmoid of z

    z can be a scalar or numpy array
    """
    return 1 / (1 + np.exp(-z))
```

```python
# Values from -10 to 10
z_tmp = np.arange(-10, 11)

# Apply sigmoid
y = sigmoid(z_tmp)

# Pretty printing
np.set_printoptions(precision=3)
print("\nInput (z)   Sigmoid(z)")
print(np.c_[z_tmp, y])
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(z_tmp, y)
ax.axvline(0, linestyle='--')   # vertical line at z = 0
ax.axhline(0.5, linestyle='--') # horizontal line at 0.5

ax.set_title("Sigmoid Function")
ax.set_xlabel("z")
ax.set_ylabel("sigmoid(z)")

plt.show()
```

```python
# Training data
x_train = np.array([0., 1, 2, 3, 4, 5])
y_train = np.array([0, 0, 0, 1, 1, 1])

# Initial parameters
w = 1.0
b = -3.0

# Model prediction
z = w * x_train + b
y_pred = sigmoid(z)

# Plot
fig, ax = plt.subplots(figsize=(6, 4))

# Data points
ax.scatter(x_train[y_train == 0], y_train[y_train == 0],
           marker='o', facecolors='none', edgecolors='blue', label='y=0')

ax.scatter(x_train[y_train == 1], y_train[y_train == 1],
           marker='x', color='red', label='y=1')

# Sigmoid curve
x_line = np.linspace(0, 5, 200)
ax.plot(x_line, sigmoid(w * x_line + b), label='Sigmoid model')

ax.set_xlabel("x")
ax.set_ylabel("Probability")
ax.set_title("Logistic Regression with Sigmoid")
ax.legend()

plt.show()
```
