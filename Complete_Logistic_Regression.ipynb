```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# multiple feature vectors
x1 = np.array([50,100,150,200,250,300])
x2 = np.array([2,3,5,4,6,7])
# Feature Matrix
X = np.column_stack((x1,x2))
#Labels
y= np.array([0,0,0,1,1,1])
#Parameters
w=np.array([-0.8,0.8])
b= -3.0
#Function
def logistic_fun(z):
    return 1/(1+np.exp(-z))
# Model Prediction
z= np.dot(X,w)+b
y_pred = logistic_fun(z)
```

```python
# Print to check
print(z)
print(y_pred)
```

```python
# Implement loss function
# We use epsilon + clip because logarithms cannot handle 0 or 1, and sigmoid can numerically reach those values.
def loss_fun(y,y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred,epsilon,1-epsilon)
    return -y*np.log(y_pred)-(1-y)*np.log(1-y_pred)

# Compute loss for each data point
loss_per_sample = loss_fun(y,y_pred)
print("Loss for each sample")
print(loss_per_sample)
```

```python
# Cost Function
cost = np.mean(loss_per_sample)
print("Average Cost:- ", cost)
```

```python
# Gradient Descent
m = len(y)
w = np.zeros(X.shape[1])
b = 0.0
alpha = 0.1
iterations = 1000
costs = []
for i in range(iterations):
    z = np.dot(X,w) + b
    y_pred = logistic_fun(z)

    cost = np.mean(loss_fun(y,y_pred))
    costs.append(cost)

    # derivative
    dw = (1/m)*np.dot(X.T,(y_pred - y))
    db = (1/m)*np.sum(y_pred-y)

    # Update Parameters
    w -= alpha*dw
    b -= alpha*db

    # Print after 100 iterations
    if i % 100 == 0:
        print(f"Iteration {i}  : Cost = {cost:.4f}")


z_final = np.dot(X, w) + b
y_pred_final = logistic_fun(z_final)
print("\n Final Weights:",w)
print("Final Bias:",b)
```

```python
# Plotting after gradient descent
plt.figure(figsize=(8,6))
# Class 0
plt.scatter(x1[y==0], x2[y==0],
            marker='o', facecolors='none',
            edgecolors='blue', label='y = 0')
# Class 1
plt.scatter(x1[y==1], x2[y==1],
            marker='x', color='red', label='y = 1')
x1_line = np.linspace(min(x1)-10, max(x1)+10, 100)
x2_line = -(w[0] * x1_line + b) / w[1]
plt.plot(x1_line, x2_line,
         color='green', linewidth=2,
         label='Decision Boundary (After Training)')
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Logistic Regression After Gradient Descent")
plt.legend()
plt.grid(True)
plt.show()
```
